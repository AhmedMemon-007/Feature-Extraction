```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.DataFrame({
    'text': ["people watch campusx","campusx a a a a watch campusx","people write comment","campusx write comment"],
    'output': [1,1,0,0]
})
```

```python
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
#OR cv=CountVectorizer(ngram_range=(1,1))
```

```python
bow=cv.fit_transform(df['text'])
```

```python
print(cv.vocabulary_) #vocabulary_  is an attribute in this class to find unique vocabulary. Each word is assigned a column index and Order is decided internally
```

```python
print(bow[0].toarray())
print(bow[1].toarray())
```

```python
cv.transform(["campusx watch and write comment of campusx"]).toarray()    #'and' and 'of' were OOV but the fact that we didn't had these in our training, here these didn't count
```

```python
bi_cv=CountVectorizer(ngram_range=(2,2)) #Bi-gram
```

```python
bigram=bi_cv.fit_transform(df['text'])
```

```python
print(bi_cv.vocabulary_)
```

```python
print(bigram[0].toarray())
```

```python
tri_cv=CountVectorizer(ngram_range=(3,3)) #Tri-gram
```

```python
trigram=tri_cv.fit_transform(df['text'])
```

```python
print(tri_cv.vocabulary_)
```

```python
print(trigram[0].toarray())
print(trigram[1].toarray())
print(trigram[2].toarray())
print(trigram[3].toarray())
```

```python
#range= 1 to 3 all combinations
n_cv=CountVectorizer(ngram_range=(1,3))
```

```python
ngram=n_cv.fit_transform(df['text'])
```

```python
print(n_cv.vocabulary_)
print(len(n_cv.vocabulary_))
```

```python
print(ngram[0].toarray())
print(ngram[1].toarray())
print(ngram[2].toarray())
print(ngram[3].toarray())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf=TfidfVectorizer()
tfidf.fit_transform(df['text']).toarray()
```

```python
print(tfidf.idf_)   #IDF of each unique word
print(tfidf.get_feature_names_out())
```
